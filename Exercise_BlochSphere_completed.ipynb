# Exercise: Understanding Qubit States

## Objective

Learn to represent and visualize basic qubit states on the Bloch sphere.

## Tasks

1. Initialize a qubit in the $|0\rangle$ state and visualize it on the Bloch sphere.
2. Apply the X gate to change the qubit to the $|1\rangle$ state and visualize it.
3. Apply the Hadamard gate to create a superposition state and visualize it.


## Import libraries

This notebook uses Qiskit to create quantum circuits, generate statevectors, and visualize states on the Bloch sphere.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt
import math
```

## Helper function


```python
def show_state(title, circuit):
    print('\n' + '=' * 60)
    print(title)
    print('=' * 60)

    state = Statevector.from_instruction(circuit)
    print('Statevector:')
    print(state)

    print('\nCircuit:')
    print(circuit.draw(output='text'))

    circuit.draw(output='mpl')
    plt.title(title + ' - Circuit')
    plt.show()

    plot_bloch_multivector(state)
    plt.title(title + ' - Bloch Sphere')
    plt.show()

    return state
```

## Part 1: Initialize a qubit in the $|0\rangle$ state


```python
qc_0 = QuantumCircuit(1)
state_0 = show_state('Part 1: Qubit initialized in |0> state', qc_0)
```

The default state of a qubit is $|0\rangle$.

On the Bloch sphere, $|0\rangle$ is located at the north pole.


## Part 2: Apply X gate to create the $|1\rangle$ state


```python
qc_1 = QuantumCircuit(1)
qc_1.x(0)

state_1 = show_state('Part 2: Apply X gate to create |1> state', qc_1)
```

The X gate changes $|0\rangle$ into $|1\rangle$.

On the Bloch sphere, $|1\rangle$ is located at the south pole.


## Part 3: Apply Hadamard gate to create a superposition


```python
qc_h = QuantumCircuit(1)
qc_h.h(0)

state_h = show_state('Part 3: Apply Hadamard gate to create superposition', qc_h)
```

The Hadamard gate changes $|0\rangle$ into:

$$\frac{|0\rangle + |1\rangle}{\sqrt{2}}$$

This means the qubit has equal probability of being measured as 0 or 1.

On the Bloch sphere, this state is shown on the equator.


## Tests


```python
# |0> should be [1, 0]
assert abs(state_0.data[0] - 1) < 1e-9
assert abs(state_0.data[1] - 0) < 1e-9

# |1> should be [0, 1]
assert abs(state_1.data[0] - 0) < 1e-9
assert abs(state_1.data[1] - 1) < 1e-9

# H|0> should be [1/sqrt(2), 1/sqrt(2)]
expected = 1 / math.sqrt(2)
assert abs(state_h.data[0] - expected) < 1e-9
assert abs(state_h.data[1] - expected) < 1e-9

print('All tests passed.')
```

## Reflection

In this exercise, I learned how to represent basic qubit states using Qiskit.
The $|0\rangle$ state is the default state of a qubit and appears at the north pole of the Bloch sphere.

After applying the X gate, the qubit changes from $|0\rangle$ to $|1\rangle$.
This is similar to a NOT operation in classical computing. On the Bloch sphere, the $|1\rangle$ state appears at the south pole.

After applying the Hadamard gate, the qubit enters a superposition state:

$$\frac{|0\rangle + |1\rangle}{\sqrt{2}}$$

This means the qubit has equal probability of being measured as 0 or 1. On the Bloch sphere, this state is shown on the equator.

The Bloch sphere visualization helps me understand that quantum states are not only 0 or 1.
They can also exist as superpositions, which is one of the key ideas in quantum computing.
