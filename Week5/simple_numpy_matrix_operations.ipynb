# Simple Matrix Operations with NumPy

## 1. Import NumPy

```python
import numpy as np
```

## 2. Create Two Matrices

Here we create two simple 2 × 2 matrices.

```python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)
```

## 3. Matrix Addition

`+` adds two matrices element by element.

```python
print("A + B:")
print(A + B)
```

## 4. Matrix Subtraction

`-` subtracts two matrices element by element.

```python
print("A - B:")
print(A - B)
```

## 5. Element-wise Multiplication

`*` multiplies matching elements.

This is **not** standard matrix multiplication.

```python
print("A * B:")
print(A * B)
```

## 6. Matrix Multiplication

`@` performs real matrix multiplication.

You can also use `np.dot(A, B)`.

```python
print("A @ B:")
print(A @ B)

print("\nnp.dot(A, B):")
print(np.dot(A, B))
```

## 7. Division

`/` divides matching elements.

This is element-wise division.

```python
print("B / A:")
print(B / A)
```

## 8. Transpose

`.T` changes rows into columns.

```python
print("A Transpose:")
print(A.T)
```

## 9. Identity Matrix

`np.eye(size)` creates an identity matrix.

```python
I = np.eye(2)

print("Identity Matrix:")
print(I)
```

## 10. Check Matrix Shape

`.shape` tells the size of the matrix.

```python
print("Shape of A:", A.shape)
print("Shape of B:", B.shape)
```

## 11. Simple User Input Version

This version lets the user enter two matrices manually.

Only normal numbers are used. No complex numbers.

```python
def input_matrix(name, rows, cols):
    matrix = []
    print(f"Enter values for Matrix {name}:")
    
    for i in range(rows):
        row = input(f"Row {i + 1}: ").split()
        row = [float(x) for x in row]
        matrix.append(row)
    
    return np.array(matrix)
```

```python
# Uncomment this section if you want manual input.

rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

user_A = input_matrix("A", rows, cols)
user_B = input_matrix("B", rows, cols)

print("\nUser Matrix A:")
print(user_A)

print("\nUser Matrix B:")
print(user_B)

print("\nA + B:")
print(user_A + user_B)

print("\nA - B:")
print(user_A - user_B)

print("\nA * B:")
print(user_A * user_B)

if cols == rows:
    print("\nA @ B:")
    print(user_A @ user_B)
else:
    print("\nMatrix multiplication A @ B is not valid for these dimensions.")
```
